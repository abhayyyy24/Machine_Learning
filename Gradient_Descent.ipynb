```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.array([1,2,3,4,5])
y=np.array([3,5,7,9,11])
```

```python
w=0
b=0

learning_rate=0.01
epochs=1000
```

```python
n=len(x)

for i in range(epochs):
  y_pred=w*x+b
  error = y_pred - y
  cost=(1/n)*np.sum(error ** 2)

  dw=(2/n)* np.sum(error*x)
  db=(2/n)*np.sum(error)

  w-=learning_rate*dw
  b-=learning_rate*db

  if i % 100==0:
    print(f"epoch {i}: cost={cost:.4f},w={w:.4f},b={b:.4f}")
```

```python
print(f"\nFinal model: y = {w:.2f}x + {b:.2f}")
```

```python
# 6. Visualize the result
plt.scatter(x, y, color='red', label='Actual data')  # plot original points
plt.plot(x, w * x + b, color='blue', label='Fitted line')  # plot regression line
plt.xlabel('X')
plt.ylabel('y')
plt.title('Linear Regression using Gradient Descent')
plt.legend()
plt.grid()
plt.show()
```
